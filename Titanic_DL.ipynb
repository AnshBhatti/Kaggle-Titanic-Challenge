```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
```

```python
train_data=pd.read_csv("train.csv")
test_data=pd.read_csv("test.csv")
```

```python
features=["Sex","Pclass","Age","SibSp","Parch"]
y=train_data["Survived"]
train_data=train_data[features]
train_data=train_data.fillna(40)
train_data["Sex"]=(train_data["Sex"]=="male")*1
x_data=torch.tensor([list(train_data.loc[each]) for each in list(train_data.index)]).float()
print(x_data)
```

```python
model=nn.Sequential(nn.Linear(5,10),nn.Linear(10,1),nn.Sigmoid())
```

```python
criterion=nn.BCELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
y=torch.tensor(y).float()
min_loss=100000000
```

```python
for epoch in range(15000):
    optimizer.zero_grad()
    o=model(x_data.float())
    loss=criterion(o,y)
    loss.backward()
    optimizer.step()
    if epoch%10000==0:
        print(epoch)
    min_loss=min(min_loss,loss.item())
```

```python
test=test_data[features].copy()
test["Sex"]=(test["Sex"]=="male")*1
test.fillna(0)
x_data=torch.tensor([list(test.loc[each]) for each in list(test.index)]).float()
```

```python
answer=model(x_data).detach().numpy()
test_data["Survived"]=(answer>=0.5)*1
test_data=test_data[["PassengerId","Survived"]]
```
